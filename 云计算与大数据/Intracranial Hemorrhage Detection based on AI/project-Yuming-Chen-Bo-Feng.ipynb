####  <font face="Times New Roman">  4.2.2 Visualization for  Different `Windows` </font>

> <font face="Times New Roman">**Window**:also known as ~grey-level mapping`, `contrast stretching`, `histogram  modification` or `contrast enhancement` is <u>the process in which the CT image greyscale component of an image is manipulated via the CT numbers; doing this will change the appearance of the picture to highlight particular structures.</u> The brightness of the image is, adjusted via the window level. The contrast is adjusted via the window width.</font>

<font face="Times New Roman">CT image values correspond to Hounsfield units (HU). 
The values stored in CT Dicoms are not Hounsfield units, but it is instead a scaled version. To extract the Hounsfield units, a  linear transformation should be applied which can be deduced from the Dicom tags.
Once the pixel values have been transformed into Hounsfield units, a windowing can be applied. The usual values for a head CT are a center of 40 and a width of 80, but we can also extract this from the Dicom headers.</font>

````python
def window_image(image, window_center,window_width, intercept, slope):
    image = (image*slope +intercept)
    image_min = window_center - window_width//2
    image_max = window_center + window_width//2
    image[image<image_min] = image_min
    image[image>image_max] = image_max
    return image

def get_windowing(data):
    dicom_fields = [data[('0028','1050')].value, #window center
                    data[('0028','1051')].value, #window width
                    data[('0028','1052')].value, #intercept
                    data[('0028','1053')].value] #slope
    def get_first_of_dicom_field_as_int(x):
        if type(x) == pydicom.multival.MultiValue:
            return int(x[0])
        else:
            return int(x)
    return [get_first_of_dicom_field_as_int(x) for x in dicom_fields]
````

````python
def view_images(images:dict,
               width:int=5,
               height:int = 2,
               size:tuple=(15,5),
               ):
  
    fig, axs = plt.subplots(height, width, figsize=size)
    
    for im in range(0, height * width):
        ID = list(images.keys())[im]
        image = images[ID][0]
        i = im // width
        j = im % width
        axs[i,j].imshow(image, cmap=plt.cm.bone) 
        axs[i,j].axis('off')
        axs[i,j].set_title(images[ID][1])
    plt.show()
````

##### <font face="Times New Roman">4.2.2.2 Brain Windowing</font>

<font face="Times New Roman">According to the Online Reference, the `Brain Windowling` set the range of values that corresponds with brain matter which will clip everything outside that range which make it more contrast in the brain-range. </font>

````python
def brain_window(dcm):
    window_min = 0
    window_max = 80
    _, _, intercept, slope = get_windowing(dcm)
    image = dcm.pixel_array
    image = image * slope + intercept
    image[image < window_min] = window_min
    image[image > window_max] = window_max
    image = (image - np.min(image)) / (np.max(image) - np.min(image))
    return image

brain_images = {instance[0]:(brain_window(read_dicom(instance[1])),lookup_label(instance[0])) for instance in dataset.take(10)}
view_images(brain_images)
````

##### <font face="Times New Roman">4.2.2.3 Three Channels Windowing</font>

<font face="Times New Roman">According to the Online Reference, the `Three Channels Windowing` use one window setting and improve the contrast by spreading it out across all 3 channels (R,G,B). </font>

````python
def all_channels_window(dcm):
    grey_image = brain_window(dcm) * 3.0
    all_chan_image = np.zeros((grey_image.shape[0], grey_image.shape[1], 3))
    all_chan_image[:, :, 0] = np.clip(grey_image - 1.0, 0.0, 1.0)
    all_chan_image[:, :, 1] = np.clip(grey_image - 2.0, 0.0, 1.0)
    all_chan_image[:, :, 2] = np.clip(grey_image, 0.0, 1.0)   
    return all_chan_image

all_channels_images = {instance[0]:(all_channels_window(read_dicom(instance[1])),lookup_label(instance[0])) for instance in dataset.take(10)}
view_images(all_channels_images)
````

##### <font face="Times New Roman">4.2.2.4 Brain + Subdural + Bone Windowing</font>

<font face="Times New Roman">The `BSB Windowing` combines `Brain Windowing`, `Subdural Windowing` and `Bone Windowing` According to the Online Reference, different hemmorhages become more obvious at these three window settings. Therefore, `BSB Windowing` may improve the  contrast  more clearly. </font>

````python
def window_image(dcm, window_center, window_width):
    _, _, intercept, slope = get_windowing(dcm)
    image = dcm.pixel_array * slope + intercept
    image_min = window_center - window_width // 2
    image_max = window_center + window_width // 2
    image[image < image_min] = image_min
    image[image > image_max] = image_max
    image = (image - np.min(image)) / (np.max(image) - np.min(image))
    return image

def bsb_window(dcm):
    brain_image = window_image(dcm, 40, 80)
    subdural_image = window_image(dcm, 80, 200)
    bone_image = window_image(dcm, 600, 2000)
    
    bsb_image = np.zeros((brain_image.shape[0], brain_image.shape[1], 3))
    bsb_image[:, :, 0] = brain_image
    bsb_image[:, :, 1] = subdural_image
    bsb_image[:, :, 2] = bone_image
    return bsb_image

bsb_images = {instance[0]:(bsb_window(read_dicom(instance[1])),lookup_label(instance[0])) for instance in dataset.take(10)}
view_images(bsb_images)
````

#### <font face="Times New Roman"> 5.2.1 Model Introduction</font>

 <font face="Times New Roman"> The model used in this project is **InceptionNetV3** which is a kind of convolution neural network  for assisting in image analysis and object detection, and got its start as a module for `Googlenet`, established in 2014. </font>

#### <font face="Times New Roman"> 5.2.2 Model Architecture</font>

<font face="Times New Roman"> The architecture of **InceptionNetV3** comes from the Inception family that makes several improvements including using Label Smoothing, Factorized 7 x 7 convolutions, and the use of an auxiliary classifer to propagate label information lower down the network (along with the use of batch normalization for layers in the sidehead).</font>
![gfJMWT.png](https://z3.ax1x.com/2021/05/18/gfJMWT.png)


#### <font face="Times New Roman">  5.2.3 Model Building </font>

<font face="Times New Roman"> The tool which is used to build the **InceptionNetV3** is `Pytorch` which is an open source machine learning framework that accelerates the path from research prototyping to production deployment.</font> 
    
    
<font face="Times New Roman"> The model is built outside the Jupyter and the code is below:</font>

---

 ```python
# in3_layers is a module contain basic layer of InceptionNet3
from in3_layers import InceptionA,InceptionB,InceptionC
from in3_layers import InceptionD,InceptionE,BasicConv2d
from collections import OrderedDict
import torch.nn.functional as F
import torch.nn as nn
import torch

INPUT_SIZE = (299,299)

class InceptionNet3(nn.Module):
    def __init__(self, num_classes=5):
        super(InceptionNet3, self).__init__()
        self.Conv2d_1a_3x3 = BasicConv2d(3, 32, kernel_size=3, stride=2)
        self.Conv2d_2a_3x3 = BasicConv2d(32, 32, kernel_size=3)
        self.Conv2d_2b_3x3 = BasicConv2d(32, 64, kernel_size=3, padding=1)
        self.Conv2d_3b_1x1 = BasicConv2d(64, 80, kernel_size=1)
        self.Conv2d_4a_3x3 = BasicConv2d(80, 192, kernel_size=3)
        self.Mixed_5b = InceptionA(192, pool_features=32)
        self.Mixed_5c = InceptionA(256, pool_features=64)
        self.Mixed_5d = InceptionA(288, pool_features=64)
        self.Mixed_6a = InceptionB(288)
        self.Mixed_6b = InceptionC(768, channels_7x7=128)
        self.Mixed_6c = InceptionC(768, channels_7x7=160)
        self.Mixed_6d = InceptionC(768, channels_7x7=160)
        self.Mixed_6e = InceptionC(768, channels_7x7=192)
        self.Mixed_7a = InceptionD(768)
        self.Mixed_7b = InceptionE(1280)
        self.Mixed_7c = InceptionE(2048)
        self.fc = nn.Linear(2048, num_classes)
        for m in self.modules():
            if isinstance(m, nn.Conv2d) or isinstance(m, nn.Linear):
                import scipy.stats as stats
                stddev = m.stddev if hasattr(m, 'stddev') else 0.1 #if else 
                X = stats.truncnorm(-2, 2, scale=stddev)
                values = torch.Tensor(X.rvs(m.weight.data.numel()))
                values = values.view(m.weight.data.size())
                m.weight.data.copy_(values)
            elif isinstance(m, nn.BatchNorm2d):
                m.weight.data.fill_(1)
                m.bias.data.zero_()


    def forward(self, x):
        x.view(-1,3,INPUT_SIZE[0],INPUT_SIZE[1])
        x = self.Conv2d_1a_3x3(x)
        x = self.Conv2d_2a_3x3(x)
        x = self.Conv2d_2b_3x3(x)
        x = F.max_pool2d(x, kernel_size=3, stride=2)
        x = self.Conv2d_3b_1x1(x)
        x = self.Conv2d_4a_3x3(x)
        x = F.max_pool2d(x, kernel_size=3, stride=2)
        x = self.Mixed_5b(x)
        x = self.Mixed_5c(x)
        x = self.Mixed_5d(x)
        x = self.Mixed_6a(x)
        x = self.Mixed_6b(x)
        x = self.Mixed_6c(x)
        x = self.Mixed_6d(x)
        x = self.Mixed_6e(x)
        x = self.Mixed_7a(x)
        x = self.Mixed_7b(x)
        x = self.Mixed_7c(x)
        x = F.avg_pool2d(x, kernel_size=8)
        x = F.dropout(x, training=self.training)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x
```

from pyspark.sql.functions import utrain
