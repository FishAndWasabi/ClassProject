```python
df[['last_updated','sonar_issues']]
```

As the above shown, the dataset is a time serie which contains data from 2011 to 2017. 

Then, we split the dataset into different parts group by `year` and clean the missing value.

```python
df['year'] = pd.to_datetime(df['last_updated']).dt.year
values = [ df.loc[df[df['year'] == year].index,'sonar_issues'].dropna() for year in df['year'].unique()]
fig,ax = plt.subplots(figsize = (15,5))
ax = creative_boxplot(ax,values,labelset=list(df['year'].unique()),
                       whiskercolor='#021211',
                       capcolor='#0B090A',
                       outliercolor='#DAFBF9',
                       outlierlinecolor='#021211',
                       boxfacecolor = '#EDFDFC',
                       curfacecolor = '#F4F9F0',
                       trendcolor='#084945',
                       mediancolor = 'black')
```

Accoring to the plot, we can find a disproportionate amount of variation in the data. Therefore, we use the `log transformation` to normalize the data.

```python
df['year'] = pd.to_datetime(df['last_updated']).dt.year
def process(value):
    value = np.log(value)
    value = value[~np.isinf(value)]
    return value
values = [process(df.loc[df[df['year'] == year].index,'sonar_issues']).dropna() for year in df['year'].unique()][::-1]

fig,ax = plt.subplots(figsize = (15,5))
ax = creative_boxplot(ax,values,labelset=list(df['year'].unique())[::-1],
                       whiskercolor='#021211',
                       capcolor='#0B090A',
                       outliercolor='#DAFBF9',
                       outlierlinecolor='#021211',
                       boxfacecolor = '#EDFDFC',
                       curfacecolor = '#F4F9F0',
                       trendcolor='#084945',
                       mediancolor = 'black')
```

**Analysis**
+ As time goes by, the number of FOSS apps increases. 
+ Based on the trends, the numbers of `sonar_issues` in each code have not changed much with the time going by, with most still concentrated on 4-6 `sonar_issues` per code.
+ As time increases,the probability distribution of `sonar_issues` gets closer to a normal distribution
